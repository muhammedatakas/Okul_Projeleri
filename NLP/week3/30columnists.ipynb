```python
import random
continuations = []
for i in range(1, 51):
    with open(f"30Columnists/raw_texts/1/1.{i}.txt", "r", encoding='latin1') as f:
        content = f.read().strip()
        continuations.append(content)

random.shuffle(continuations)

sentences = [cont for cont in continuations]

print(*sentences, sep="\n")
```

```python
import re
from nltk import word_tokenize
import nltk
import string

nltk.download('punkt')

# Function to remove punctuation and special characters from tokens
def remove_punctuation(tokens):
    # Define a regular expression pattern to match unwanted characters
    pattern = re.compile(r'[^a-zA-Z0-9<>\s]')
    return [pattern.sub('', token) for token in tokens if pattern.sub('', token)]


# Tokenize sentences and add <bos> and <eos>
sentence_tokens = []
for sentence in sentences:
    tokens = ["<bos>"] + [token.lower() for token in word_tokenize(sentence)] + ["<eos>"]
    tokens = remove_punctuation(tokens)
    sentence_tokens.append(tokens)

# Flatten the list of sentence tokens
token_list = [x for X in sentence_tokens for x in X]

# Print the tokenized sentences
print(*sentence_tokens, sep="\n")
```

```python
import re
from nltk.tokenize import casual
import nltk
import string

nltk.download('punkt')

# Function to remove unwanted punctuation while preserving contractions
def remove_unwanted_punctuation(tokens):
    # Define a regular expression pattern to match unwanted characters
    pattern = re.compile(r'[^a-zA-Z0-9<>\s\']')
    return [pattern.sub('', token) for token in tokens if pattern.sub('', token)]


# Tokenize sentences and add <bos> and <eos>
sentence_tokens = []
for sentence in sentences:
    tokens = ["<bos>"] + [token.lower() for token in casual.casual_tokenize(sentence)] + ["<eos>"]
    tokens = remove_unwanted_punctuation(tokens)
    sentence_tokens.append(tokens)

# Flatten the list of sentence tokens
token_list = [x for X in sentence_tokens for x in X]

# Print the tokenized sentences
print(*sentence_tokens, sep="\n")
```

```python
# 3-gram frequencies
threegrams = []
for sentence in sentence_tokens:
  for i in range(len(sentence)-2):
    threegrams.append((sentence[i], sentence[i+1], sentence[i+2]))

threegram_freqs = {x:0 for x in threegrams};
for threegram in threegrams:
  threegram_freqs[threegram] += 1;

freqs = [(x,threegram_freqs[x]) for x in threegram_freqs]; # Convert the fdist to list
freqs.sort(key=lambda t: t[1], reverse=1);

print("3-gram freqs:", *freqs, sep="\n")
```

```python
# 3-gram sentence generator function
import random as rnd

def generate_next_3grams(context):
  curr_token_1, curr_token_2 = context[-2], context[-1]
  possible_threegrams = [threegram for threegram in threegrams if threegram[0] == curr_token_1 and threegram[1] == curr_token_2]


  if len(possible_threegrams) == 0:
    return "<eos>"

  sample_space = []
  for threegram in possible_threegrams:
    sample_space.append([threegram]*threegram_freqs[threegram])

  sample_space = [x for X in sample_space for x in X]

  next = rnd.choice(sample_space)

  return next[2]
```

```python
sentence = ["<bos>","it"]

while True:
  next_token = generate_next_3grams(sentence).replace(' ', '')
  sentence.append(next_token)
  if next_token == "<eos>" or len(sentence) > 10:
    break

print(sentence)
```
